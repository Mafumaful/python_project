# 使用opencv提取照片并用matplotlib将照片显示出来

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# 读取图像
img = cv2.imread('./images/imgLena.png')
# 显示图像
plt.imshow(img)
```

由于opencv提取的颜色格式是BGR格式，而matplotlib根据RGB颜色格式显示颜色，所以我们需要在读取图像以后对图像进行颜色格式的转化

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```
