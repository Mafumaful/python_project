# 使用opencv提取照片并用matplotlib将照片显示出来

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# 读取图像
img = cv2.imread('./images/imgLena.png')
# 显示图像
plt.imshow(img)
plt.axis('off')
```

由于opencv提取的颜色格式是BGR格式，而matplotlib根据RGB颜色格式显示颜色，所以我们需要在读取图像以后对图像进行颜色格式的转化

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.axis('off')
```

同样的，我们也可以用matplotlib里面内置的函数来读取图像

```python
import matplotlib.image as mpimg

# Read the image file
image = mpimg.imread('./images/imgLena.png')

# Display the image
plt.imshow(image)
plt.axis('off')
```
