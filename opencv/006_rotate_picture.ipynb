# 图片旋转

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image = cv2.imread("./images/cat_dog.jpeg")
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
shape = image.shape
plt.imshow(image)
plt.axis('off')
```

```python
# 旋转
Rot = cv2.getRotationMatrix2D((shape[1] / 2.0, shape[0] / 2.0), 9.4, 1)           # 旋转的大小自己定的
# 
result = cv2.warpAffine(image, Rot, (shape[1], shape[0]),                       #默认差值运算
                    borderValue=(125, 125, 125))    
dst1 = cv2.warpAffine(image, Rot, (shape[1], shape[0]), flags=cv2.INTER_NEAREST,
                        borderValue=(125, 125, 125))                            # 使用最近临值法
dst2 = cv2.warpAffine(image, Rot, (shape[1], shape[0]), flags=cv2.INTER_LINEAR,
                        borderValue=(125, 125, 125))                            # 使用双线性法
```

```python
plt.subplot(2,2,1)
plt.title('(a) raw')
plt.imshow(image)      #设置区域并画图
plt.axis('off')

plt.subplot(2,2,2)
plt.title('(b) default')
plt.imshow(result)
plt.axis('off')

plt.subplot(2,2,3)
plt.title('(c) inter nearest')
plt.imshow(dst1)
plt.axis('off')

plt.subplot(2,2,4)
plt.title('(d) inter linear')
plt.imshow(dst2)
plt.axis('off')

plt.show()
```
